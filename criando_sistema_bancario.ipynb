```python
# O projeto trata-se de um sistema bancario simplificado, onde o cliente pode acessar seu extrato, depositar e sacar dinheiro

# Menu principal, sera sempre a tela que o cliente ira receber ao terminar uma transacao

menu = """

Olá! É uma enorme satisfação ter você aqui conosco!

Qual ação gostaria de fazer?

[d] Depositar
[s] Sacar
[e] Extrato
[q] Sair

=> """

# Declaracao de variaveis
saldo = 0
limite = 500
extrato = f'''

##### EXTRATO #####

Saldo inicial de R$ {saldo:.2f}\n
'''
numero_saques = 0
LIMITE_SAQUES = 3


# Laco de repeticao para manter o menu enquanto o cliente esta ativo

while True:

    opcao = input(menu)

    # Caso o cliente queira realizar um deposito
    if opcao == 'd':
        print('Sua opção selecionada foi Depósito')
        print(f'Seu saldo atual é de R$ {saldo:.2f}')
        deposito = float(input('Qual valor gostaria de depositar?'))
        confirmacao = input(f'Confirme o depósito de R$ {deposito:.2f}. s/n?') # Confirmacao para depositar o valor desejado
        if confirmacao == 's' and deposito > 0:
            saldo += deposito
            print('Depósito realizado com sucesso!\n')
            print(f'Seu novo saldo agora é de R$ {saldo:.2f}')
            extrato += f'Depósito no valor de R$ {deposito:.2f}\n'
        else:
            print('Tente a operação novamente!')
    

    # Caso o cliente queira realizar um saque
    elif opcao == 's':
        print('Sua opção selecionada foi Saque')
        if numero_saques == LIMITE_SAQUES:
            print(f'Infelizmente você atingiu seu limite de saques por hoje. Para sua segurança, pedimos a gentileza em retornar amanhã')
        else:
            print(f'Seu saldo atual é de R$ {saldo:.2f}')
            saque = int(input('Qual o valor gostaria de sacar?'))
            confirmacao = input(f'Confirme o saque de R$ {saque:.2f}. s/n?') # Confirmacao para sacar o valor desejado
            if saque <= saldo and saque <= limite and saque > 0 and confirmacao == 's': # Verificacao se o cliente nao ultrapassou o limite de saques do dia
                saldo -= saque
                print('Saque realizado com sucesso!\n')
                print(f'Seu novo saldo agora é de R$ {saldo:.2f}')
                extrato += f'Saque no valor de R$ {saque:.2f}\n'
                numero_saques += 1
            elif saque > saldo and confirmacao == 's':
                print('Saldo insuficiente')
            elif saque > limite and confirmacao == 's':
                print(f'Seu limite por saque é de R$ {limite}, tente a operação novamente!')
            else:
                print('Tente a operação novamente!')


    # Caso o cliente queira ver seu extrato com todas as motimentacoes realizadas
    elif opcao == 'e':
        print(extrato)
        print(f'Saldo atual de R$ {saldo:.2f}')
    

    # Sair do sistema
    elif opcao == 'q':
        break
    
    else:
        print('Operação inválida, por favor selecione novamente a operação desejada.')
```
